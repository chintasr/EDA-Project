### Importing  libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the data set

```python
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\project\Restaurants.csv")
df
```

```python
def categorize_rating(Rating):  # add column
    
    if Rating <= 2:
        return "Poor"
    
    elif Rating <= 3.5:
        return "Avg"
    
    elif Rating <= 4:
        return "Good"
    
    else:
        return "Excellent"
    
df["categorize_rating"]  = df["Rating"].apply(categorize_rating)
df
```

```python
df.drop("Unnamed: 0",axis = 1,inplace = True)  # drop unnamed  not recommended
```

### Data cleaning

#### Structural Errors

```python
df["Rating"].unique()  # check the structural errors
```

```python
df["Price for 2(approx)"].unique()  # check the structural errors
```

#### Missing values

```python
df.isna().sum()  # missing values
```

```python
df["Area"].fillna("Secunderabad",inplace = True)
```

```python
df["Cuisine"].fillna("North Indian,Chinese",inplace = True)
```

```python
df['Rating']=df.groupby('Restaurant_Name')['Rating'].transform(lambda x : x.fillna(round(x.mean(),1)))
```

```python
df.dropna(axis = 0,inplace = True)
```

```python
df.isna().sum()
```

#### Type Casting

```python
df.dtypes    # data type
```

```python
df["Price for 2(approx)"] = df["Price for 2(approx)"].astype("int32")
```

```python
df.dtypes
```

#### Duplicates

```python
df[df.duplicated()]  # duplicates
```

```python
df.drop_duplicates(inplace = True) # removing duplicates
```
