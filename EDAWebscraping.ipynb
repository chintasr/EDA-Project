```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np


Restaurant_Names = []    
Rating = []
Locations = [] 
Cuisine = []
Price = []
Area = []
Region = []


for i in range(1,50):
    url = f"https://www.dineout.co.in/hyderabad-restaurants?loc=Hyderabad&p={i}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text) 
    
    
    for i in soup.find_all('div', class_="restnt-main-wrap clearfix"):  #restaurant_names-->1
        name = i.find('a', class_="restnt-name").text
        if name:
            Restaurant_Names.append(name)
        else:
            Restaurant_Names.append(np.nan)  
            
            
            
    for i in soup.find_all('div', class_="restnt-main-wrap clearfix"):  # cuisine -->2
        text = i.find("span", class_="double-line-ellipsis").text
    
        regex = re.findall(r'\|(.+)',text)
        if regex:
            Cuisine.append(regex[0])
        else:
            Cuisine.append(np.nan)  
            
            
            
    
    for i in soup.find_all('div',class_="restnt-main-wrap clearfix"):  # rating--3
        r = i.find("div", class_ = "img-wrap").text
        
        regex = re.search(r'(\d+(\.\d+)?)', r)

        if regex:
            Rating.append(float(regex.group(1)))
        else:
            Rating.append(np.nan)

  
            
            
    for i in soup.find_all('div', class_="restnt-main-wrap clearfix"):  # location --->4
        loc = i.find("div",class_="restnt-loc").text
    
        regex = re.match(r'([^,]+)',loc)
    
        if regex:
            result = regex.group(1)
            Locations.append(result)
        else:
            Locations.append(np.nan)
            
            
            
    for i in soup.find_all('div', class_="restnt-main-wrap clearfix"):  # area ----> 5
        loc = i.find("div", class_="restnt-loc").text
    
    
        regex = re.search(r',\s*(.+)\s*,',loc)
    
        if regex:
            a = regex.group(1).strip()
            Area.append(a)
        else:
            Area.append(np.nan)
            
            
            
    for i in soup.find_all('div', class_="restnt-main-wrap clearfix"):  # region ---> 6
        loc = i.find("div", class_="restnt-loc").text
        location = loc.split(',')
    
        if len(location) > 1:
            last = location[-1].strip()
            Region.append(last)
        else:
            print("np.nan")  
            
            
    for i in soup.find_all('div', class_="restnt-main-wrap clearfix"):   # price--->7
        text = i.find("span", class_="double-line-ellipsis").text.replace('₹','')
        t = text.replace(",",'')
    
        regex = re.search(r'(.*)(\sfor\s)', t)

        if regex:
            result = regex.group(1).strip()
            Price.append(result)
        else:
            Price.append(np.nan)
```

```python
Restaurant = {"Restaurant_Name" :Restaurant_Names,"Cuisine":Cuisine,"Price for 2(approx)":Price,"Rating" :Rating,"Location":Locations,"Area":Area,"Region":Region,}
```

```python
df = pd.DataFrame(Restaurant)
```
